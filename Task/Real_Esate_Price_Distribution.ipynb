# Real Estate Price :
Analyze and visualize the distribution of real estate prices in different neighborhoods of a city. Use a dataset of real estate listings and create:

- A histogram to show the price distribution.
-A box plot to compare price distributions across different neighborhoods.

- Scatter plots to show the relationship between property size and price.
**Hint:** Use `plt.hist()` for histograms, `plt.boxplot()` for box plots, and `plt.scatter()` for scatter plots.

**Expectation:** Visualizations showing the distribution of real estate prices, comparisons across neighborhoods, and the relationship between property size and price.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("AmesHousing.csv")

# creating Histogram
plt.figure(figsize=(12, 6))
sns.histplot(data['SalePrice'], kde=True, bins=30, color='skyblue', edgecolor='black')
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.title("Distribution of Sale Prices")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Box Plot compare Sale Price Distribution across Neighborhoods
plt.figure(figsize=(16, 10))
sns.boxplot(x='Neighborhood', y='SalePrice', data=data, palette="Set2", hue='Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Sale Price')
plt.title('Sale Price Distribution Across Neighborhoods')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Scatter Plot with Trend Line
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Gr Liv Area', y='SalePrice', data=data, alpha=0.6, edgecolor='w', linewidth=0.5, color='dodgerblue')
sns.regplot(x='Gr Liv Area', y='SalePrice', data=data, scatter=False, color='red', line_kws={"linewidth":2})
plt.xlabel("Ground Living Area (sq ft)")
plt.ylabel("Sale Price")
plt.title("Relationship Between Property Size and Sale Price")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
